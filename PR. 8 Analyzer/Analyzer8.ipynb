```python
import numpy as np

current_numpy_array = None

class EXIT_STATUS:
    CONTINUE = 0
    GO_BACK = 1
    GO_TO_MAIN = 2
    BREAK_PROGRAM = 3

def otherArrayOperations(arr):
  def indexing(arr):
    while True:
              print('1. 1D Array')
              print('2. 2D Array')
              print('3. 3D Array')
              print('4. Go back')

              try:
                  index_choice = int(input('Enter your choice:- '))

                  if index_choice == 1:

                      i = int(input('Enter the index:- '))
                      print(arr[i])

                  elif index_choice == 2:

                      r = int(input('Enter the row index:- '))
                      c = int(input('Enter the column index:- '))
                      print(arr[r, c])

                  elif index_choice == 3:

                      d = int(input("Enter depth index:- "))
                      r = int(input("Enter row index:- "))
                      c = int(input("Enter column index:- "))
                      print(arr[d][r][c])

                  elif index_choice == 4:

                      return EXIT_STATUS.GO_BACK

                  elif index_choice == 5:
                      return EXIT_STATUS.GO_TO_MAIN

                  else:
                      print('Invalid choice')
              except ValueError:
                  print('Invalid input. Please enter a number.')
              except IndexError:
                  print('Index out of bounds. Please enter a valid index.')
    return EXIT_STATUS.CONTINUE

  def slicing(arr):
    while True:
      print('1. 1D Array')
      print('2. 2D Array')
      print('3. 3D Array')
      print('4. Go back')
      print('5. Go to Main Menu')
      try:
          slice_choice = int(input('Enter your choice:- '))

          if slice_choice == 1:


            sr = int(input('Enter the start range:- '))
            er = int(input('Enter the end range:- '))

            print(arr[sr:er])

          elif slice_choice == 2 or slice_choice == 3:

            r1 = int(input('Enter the start range of row:- '))
            r2 = int(input('Enter the end range of row:- '))
            c1 = int(input('Enter the start range of column:- '))
            c2 = int(input('Enter the end range of column:- '))

            print(arr[r1:r2,c1:c2])

          elif slice_choice == 4:

            return EXIT_STATUS.GO_BACK

          elif slice_choice == 5:
            return EXIT_STATUS.GO_TO_MAIN

          else:
            print('Invalid choice')
      except ValueError:
          print('Invalid input. Please enter a number.')
      except IndexError:
          print('Slice range out of bounds. Please enter valid ranges.')
    return EXIT_STATUS.CONTINUE


  print('====================================')

  print('Choose an operaion:')

  while True:
    print('1. Indexing')
    print('2. Slicing')
    print('3. Go back')
    print('4. Go to Main Menu')

    try:
        opr_choice = int(input('Enter your choice:- '))
        status = EXIT_STATUS.CONTINUE

        match opr_choice:
            case 1:
              status = indexing(arr)
            case 2:
              status = slicing(arr)
            case 3:
              return EXIT_STATUS.GO_BACK
            case 4:
              return EXIT_STATUS.GO_TO_MAIN
            case _:
              print('Invalid choice')

        if status != EXIT_STATUS.CONTINUE:
            return status
    except ValueError:
        print('Invalid input. Please enter a number.')
  return EXIT_STATUS.CONTINUE

def mathematicalOperations(arr):
    while True:
        print('====================================')
        print('Choose a mathematical operation:')
        print('1. Addition')
        print('2. Subtraction')
        print('3. Multiplication')
        print('4. Division')
        print('5. Go back')
        print('6. Go to Main Menu')

        try:
            math_choice = int(input('Enter your choice:- '))

            match math_choice:
                case 1:
                    scalar = input('Enter the same size array to add(elements seperated by space):- ')
                    ele_list = [int(e.strip()) for e in scalar.split(" ")]
                    scl_list = np.array(ele_list).reshape(arr.shape)
                    result = arr + scl_list
                    print('Original array:')
                    print(arr)
                    print('Second array:')
                    print(scl_list)
                    print('Result of Addition:')
                    print(result)
                case 2:
                    scalar = input('Enter the same size array to substract(elements seperated by space):- ')
                    sub_list = [int(e.strip()) for e in scalar.split(" ")]
                    scl_list = np.array(sub_list).reshape(arr.shape)
                    result = arr - scl_list
                    print('Original array:')
                    print(arr)
                    print('Second array:')
                    print(scl_list)
                    print('Result of Subtraction:')
                    print(result)
                case 3:
                    scalar = input('Enter the same size array to multiply by(elements seperated by space):- ')
                    mul_list = [int(e.strip()) for e in scalar.split(" ")]
                    scl_list = np.array(mul_list).reshape(arr.shape)
                    result = arr * scl_list
                    print('Original array:')
                    print(arr)
                    print('Second array:')
                    print(scl_list)
                    print('Result of Multiplication:')
                    print(result)
                case 4:
                    scalar = input('Enter the same size array to divison by(elements seperated by space):- ')
                    div_list = [int(e.strip()) for e in scalar.split(" ")]
                    scl_list = np.array(div_list).reshape(arr.shape)
                    if 0 not in scl_list:
                        result = arr / scl_list
                        print('Original array:')
                        print(arr)
                        print('Second array:')
                        print(scl_list)
                        print('Result of Division:')
                        print(result)
                    else:
                        print('Error: Division by zero is not allowed.')
                case 5:
                    return EXIT_STATUS.GO_BACK
                case 6:
                    return EXIT_STATUS.GO_TO_MAIN
                case _:
                    print('Invalid choice')
        except ValueError:
            print('Invalid input. Please enter a number.')
        except Exception as e:
            print(f'An error occurred: {e}')
    return EXIT_STATUS.CONTINUE


def combine_split_operaions(arr):
  while True:
        print('====================================')
        print('Choose an operation:')
        print('1. Combine Arrays')
        print('2. Split Array')
        print('3. Go back')
        print('4. Go to Main Menu')

        try:

          csa_choice = int(input('Enter your choice:- '))

          match csa_choice:
              case 1:
                  elements = list(map(int, input("\nEnter the elements of another array to combine (6 elements separated by space): ").split()))
                  arr2 = np.array(elements).reshape(arr.shape)
                  print("\nOriginal Array:")
                  print(arr)
                  print("\nSecond Array:")
                  print(arr2)
                  combined = np.vstack((arr, arr2))
                  print("\nCombined Array (Vertical Stack):")
                  print(combined)

              case 2:
                  print("\nOriginal Array:")
                  print(arr)

                  split_arr = np.array_split(arr,2)

                  print("\nSplit Arrays:")
                  for i in split_arr:
                    print(i)
              case 3:
                  return EXIT_STATUS.GO_BACK
              case 4:
                  return EXIT_STATUS.GO_TO_MAIN
              case _:
                  print('Invalid choice')

        except ValueError:
            print('Invalid input.')
        except Exception as e:
            print(f'An error occurred: {e}')
  return EXIT_STATUS.CONTINUE


def search_sort_filter_operations(arr):
    while True:
        print('====================================')
        print('Choose an operation:')
        print('1. Search value')
        print('2. Sort value')
        print('3. Filter value')
        print('4. Go back')
        print('5. Go to Main Menu')


        choice = int(input('Enter your choice:- '))

        match choice:
            case 1:
              val = int(input('Enter the value you want to seach from array:- '))
              index = np.where(arr == val)
              if len(index[0]) > 0:
                  print(f'Index of {val} is:-  {index[0][0]}')
              else:
                  print(f'{val} not found in the array.')
            case 2:
              print(np.sort(arr))
            case 3:
              filt_arr = arr[arr % 2 == 0]
              print(f'Filter array:- {filt_arr}')
            case 4:
              return EXIT_STATUS.GO_BACK
            case 5:
              return EXIT_STATUS.GO_TO_MAIN
            case _:
              print('Invalid choice')
    return EXIT_STATUS.CONTINUE


def agreegate_operations(arr):

    while True:
        print("Choose an aggregate/statistical operation:")
        print("1. Sum")
        print("2. Mean")
        print("3. Median")
        print("4. Standard Deviation")
        print("5. Variance")
        print("6. Go back")
        print("7. Go to Main Menu")

        choice = int(input("Enter your choice: "))

        print("\nArray:")
        print(arr)

        match choice:

            case 1:
              print("\nSum =", np.sum(arr))

            case 2:
              print("\nMean =", np.mean(arr))

            case 3:
              print("\nMedian =", np.median(arr))

            case 4:
              print("\nStandard Deviation =", np.std(arr))

            case 5:
              print("\nVariance =", np.var(arr))

            case 6:
              return EXIT_STATUS.GO_BACK

            case 7:
              return EXIT_STATUS.GO_TO_MAIN

            case _:
              print("Invalid Choice")
    return EXIT_STATUS.CONTINUE


def oneDArray():
    global current_numpy_array
    elements_str = input('Enter elements for 1D array (comma-separated):- ')
    try:
        elements = [int(x.strip()) for x in elements_str.split(',')]
        arr = np.array(elements)
        current_numpy_array = arr
        print('Array Created Successfully:')
        print(arr)
        status = otherArrayOperations(arr)
        return status
    except ValueError:
        print('Invalid input. Please enter numbers separated by commas.')
    return EXIT_STATUS.CONTINUE


def twoDArray():
    global current_numpy_array
    rows = int(input('Enter the number of rows:- '))
    cols = int(input('Enter the number of columns:- '))
    elements_str = input(f'Enter {rows * cols} elements for 2D array (comma-separated):- ')
    try:
        elements = [int(x.strip()) for x in elements_str.split(',')]
        if len(elements) == rows * cols:
            arr = np.array(elements).reshape(rows, cols)
            current_numpy_array = arr
            print('Array Created Successfully:')
            print(arr)
            status = otherArrayOperations(arr)
            return status
        else:
            print(f'Invalid number of elements. Expected {rows * cols}, got {len(elements)}.')
    except ValueError:
        print('Invalid input. Please enter numbers separated by commas.')
    return EXIT_STATUS.CONTINUE


def threeDArray():
    global current_numpy_array
    depth = int(input('Enter the depth (number of matrices):- '))
    rows = int(input('Enter the number of rows per matrix:- '))
    cols = int(input('Enter the number of columns per matrix:- '))
    elements_str = input(f'Enter {depth * rows * cols} elements for 3D array (comma-separated):- ')
    try:
        elements = [int(x.strip()) for x in elements_str.split(',')]
        if len(elements) == depth * rows * cols:
            arr = np.array(elements).reshape(depth, rows, cols)
            current_numpy_array = arr
            print('Array Created Successfully:')
            print(arr)
            status = otherArrayOperations(arr)
            return status
        else:
            print(f'Invalid number of elements. Expected {depth * rows * cols}, got {len(elements)}.')
    except ValueError:
        print('Invalid input. Please enter numbers separated by commas.')
    return EXIT_STATUS.CONTINUE


print('Welcome to the Numpy Analyzer!')

while True:
    print('====================================')
    print('Choose an option:')

    print('1. Create a Numpy Array')
    print('2. Perform Mathematical Operations')
    print('3. Combine or Split Arrays')
    print('4. Search, Sort or Filter Arrays')
    print('5. Compute Aggregates and Statistics')
    print('6. Exit')

    try:
        choice = int(input('Enter your choice:- '))
        status = EXIT_STATUS.CONTINUE

        match choice:
            case 1:
                while True:
                    print('====================================')
                    print('Select the type of array to create:')
                    print('1. 1D Array')
                    print('2. 2D Array')
                    print('3. 3D Array')
                    print('4. Go back')

                    try:
                        array_type = int(input('Enter your choice:- '))
                        inner_status = EXIT_STATUS.CONTINUE

                        match array_type:
                            case 1:
                                inner_status = oneDArray()
                            case 2:
                                inner_status = twoDArray()
                            case 3:
                                inner_status = threeDArray()
                            case 4:
                                inner_status = EXIT_STATUS.GO_BACK
                            case _:
                                print('Invalid choice')

                        if inner_status == EXIT_STATUS.GO_TO_MAIN:
                            status = EXIT_STATUS.GO_TO_MAIN
                            break
                        elif inner_status == EXIT_STATUS.GO_BACK:
                            status = EXIT_STATUS.GO_BACK
                            break

                    except ValueError:
                        print('Invalid input. Please enter a number.')
                if status == EXIT_STATUS.GO_TO_MAIN:
                    continue

            case 2:
                if current_numpy_array is not None:
                    status = mathematicalOperations(current_numpy_array)
                else:
                    print('No array has been created yet. Please create an array first (Option 1).')
            case 3:
                if current_numpy_array is not None:
                    status = combine_split_operaions(current_numpy_array)
                else:
                    print('No array has been created yet. Please create an array first (Option 1).')
            case 4:
                if current_numpy_array is not None:
                    status = search_sort_filter_operations(current_numpy_array)
                else:
                    print('No array has been created yet. Please create an array first (Option 1).')
            case 5:
                if current_numpy_array is not None:
                    status = agreegate_operations(current_numpy_array)
                else:
                    print('No array has been created yet. Please create an array first (Option 1).')
            case 6:
                status = EXIT_STATUS.BREAK_PROGRAM
                break

            case _:
                print('Invalid choice')

        if status == EXIT_STATUS.GO_TO_MAIN:
            continue
        elif status == EXIT_STATUS.BREAK_PROGRAM:
            break

    except ValueError:

        print('Invalid input. Please enter a number.')
```
